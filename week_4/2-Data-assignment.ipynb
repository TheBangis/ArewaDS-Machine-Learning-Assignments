```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pumpkins = pd.read_csv('../week_4/US-pumpkins.csv')
pumpkins.head()
```

```python
pumpkins.isnull().sum()
```

```python
# select only columns you want to use.

new_columns = ['Low Price', 'High Price']
pumpkins = pumpkins.drop([c for c in pumpkins.columns if c not in new_columns], axis=1)

# copy new data into data frame
new_pumpkins = pd.DataFrame({'Low_Price': pumpkins['Low Price'], 'High_Price': pumpkins['High Price']}) 


low_price = new_pumpkins.Low_Price
high_price = new_pumpkins.High_Price
# plot graph with matplotlib
plt.scatter(low_price, high_price)
plt.title('Relationship  between Low Price and High Price')
plt.show()
```
